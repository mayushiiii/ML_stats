```python
# implementare chi-squared
# testam diverse perechi de atribute
from scipy import stats
significance_matrix_discrete = pd.DataFrame(index=diabetes_discrete_vals.columns, columns=diabetes_discrete_vals.columns)
for attribute1 in diabetes_discrete_vals.columns:
    for attribute2 in diabetes_discrete_vals.columns:
        if attribute1 != attribute2:
            diabetes_crosstab = pd.crosstab(diabetes_discrete_vals[attribute1], diabetes_discrete_vals[attribute2])
            # print(diabetes_crosstab)
            diabetes_chisq = sp.stats.chi2_contingency(diabetes_crosstab)
            if(diabetes_chisq[1] < 0.05):
                print(f'Chi-squared test for {attribute1} and {attribute2} is significant: {diabetes_chisq[1]}\n')
                significance_matrix_discrete.loc[attribute1, attribute2] = diabetes_chisq[1]
            else:
                print(f'Chi-squared test for {attribute1} and {attribute2} is not significant: {diabetes_chisq[1]}\n')
                significance_matrix_discrete.loc[attribute1, attribute2] = diabetes_chisq[1]
```

#### Calcul IQR pentru determinarea de outliere

```python
# Q1 , Q3
Q1 = diabetes_continuous_vals.quantile(0.25)
Q3 = diabetes_continuous_vals.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
print("\n")
threshold = 1.5 

for attribute in diabetes_continuous_vals.columns:
    lower_bound = Q1[attribute] - threshold * IQR[attribute]
    upper_bound = Q3[attribute] + threshold * IQR[attribute]
    print(f'{attribute}: Lower bound: {lower_bound}, Upper bound: {upper_bound}')
print("\n")
# identificarea outlier-ilor
outliers = pd.DataFrame()
for attribute in diabetes_continuous_vals.columns:
    outlier_condition = (diabetes_continuous_vals[attribute] < (Q1[attribute] - threshold * IQR[attribute])) | (diabetes_continuous_vals[attribute] > (Q3[attribute] + threshold * IQR[attribute]))
    outliers[attribute] = diabetes_continuous_vals[attribute][outlier_condition]
    print(f'Number of outliers for {attribute}: {outliers[attribute].count()}')

print("\n")
```

#### Atribute redundante
Am facut o matrice in care sa se tina cont de corelatia intre atribute. Unde am considerat atributele ca fiind corelate am eliminat acele coloane din dataframe.

```python
# am vazut ca am niste nan-uri in matrice deci le voi presupune ca fiind false
significance_matrix_cont_diabetes = significance_matrix_cont_diabetes.fillna(0)
# si inlocuim false-urile cu 0
significance_matrix_cont = significance_matrix_cont_diabetes.replace(False, 0)
# eliminam redundantele, atributele categorice corelate
# deci unde avem o corelatie mai mare de 0.5
for attribute1 in diabetes_continuous_vals.columns:
    corr = 0
    for attribute2 in diabetes_continuous_vals.columns:
        if corr == 1:
            break
        if attribute1 != attribute2:
            if significance_matrix_cont_diabetes.loc[attribute1, attribute2] > 0.5:
                print(f'{attribute1} and {attribute2} are correlated')
                diabetes_continuous_vals.drop(columns=attribute2, inplace=True)
    if corr == 1:
        break            
print(diabetes_continuous_vals)
                
```
